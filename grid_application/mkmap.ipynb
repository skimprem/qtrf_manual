```python
import pygmt
import rioxarray



dataset = rioxarray.open_rasterio('./QazTRF_20250303_kz.tif')

lat = dataset.sel(band=1)  # Первый канал
lon = dataset.sel(band=2)  # Второй канал

info = pygmt.grdinfo(
    grid = lat,
    per_column = True
)

region = [float(x) for x in info.split()[:4]]

lat_fig = pygmt.Figure()

lat_fig.coast(
    region = region,
    projection = 'M8i',
    # shorelines = True,
    # water = 'lightblue',
    land = 'lightgrey',
)

lat_fig.grdimage(
    grid = lat,
    frame = True,
    cmap = 'turbo',
    shading = True,
    # transparency = 50,
    # nan_transparent = True,
)

lat_fig.coast(
    # shorelines = True,
    water = 'lightblue',
    # land = 'lightbrown',
    frame = 'ag',
)

lat_fig.colorbar(
    frame=['x+lСдвиг', 'y+lУгл. сек.']
)

lat_fig.savefig('figs/lat.pdf')

lon_fig = pygmt.Figure()

lon_fig.coast(
    region = region,
    projection = 'M8i',
    # shorelines = True,
    # water = 'lightblue',
    land = 'lightgrey',
)

lon_fig.grdimage(
    grid = lon,
    frame = True,
    cmap = 'turbo',
    shading = True,
    # transparency = 50,
    # nan_transparent = True,
)

lon_fig.coast(
    # shorelines = True,
    water = 'lightblue',
    # land = 'lightbrown',
    frame = 'ag',
)

lon_fig.colorbar(
    frame=['x+lСдвиг', 'y+lУгл. сек.']
)

lon_fig.savefig('figs/lon.pdf')
```
